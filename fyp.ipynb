```python
numerical_variables = ['age', 'grades'] #contains numerical variable in dataset to create histogram
colors = ['#ADD8E6', '#FFB6C1'] #set the colours for histogram, example ADD8E6 for 'age' and FFB6C1 for 'grades'

for column, color in zip(numerical_variables, colors):
    plt.figure(figsize=(8, 5)) #sets figure size
    sns.histplot(data=df, x = column, kde=True, color = color) #insert the dataset, create histogram and colour,as for ''kde='true' it means include a kernel density estimate curve
    plt.title(f'Distribution of {column}') #uses f-string to dynamically insert the variable name into the title
    plt.show() #display histogram plot
```
